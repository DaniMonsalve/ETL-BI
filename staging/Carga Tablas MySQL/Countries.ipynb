Nos damos cuenta de que el proceso no es eficiente y optamos por usar una libreria para que pasandole como input los nombres de los paises, nos matchee los codigos de dos y tres digitos

```python
input_countries = list(table['name'])
countries = {}
for country in pycountry.countries:
    countries[country.name] = country.alpha_3

codes = [countries.get(country, 'NaN') for country in input_countries]
```

```python
table['country_code'] = codes
```
